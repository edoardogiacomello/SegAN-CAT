# Examples
This is an example for Brats2019 "v2". v1 is the one used in the Segan_CAT paper. The v2 has 3-fold splitted to have a separate test set.

```python
import dataset_helpers as dh
import glob
import pandas as pd
```

```python
log = glob.glob('../../datasets/RAW/BRATS2019/MICCAI_BraTS_2019_Data_Training/*/*/')
pd_paths = pd.DataFrame({'path':[bt for bt in log], 'patient': [bt.split('_')[-2] for bt in log], 'dataset_version': [bt.split('_')[-3] for bt in log], 'grade': [bt.split('/')[-3] for bt in log]})
```

```python
train_split = pd.DataFrame()
valid_split = pd.DataFrame() 
test_split = pd.DataFrame()
train_ratio = 0.7
valid_test_balance = 0.5
valid_ratio =  train_ratio + (1 - train_ratio)*valid_test_balance
```

```python
for g, group in pd_paths.groupby(['grade', 'dataset_version']):
    shuffled = group.sample(frac=1)
    train_split = train_split.append(shuffled.iloc[:int(train_ratio*len(shuffled))], ignore_index=True)
    valid_split = valid_split.append(shuffled.iloc[int(train_ratio*len(shuffled)):int(valid_ratio*len(shuffled))], ignore_index=True)
    test_split = test_split.append(shuffled.iloc[int(valid_ratio*len(shuffled)):], ignore_index=True)
```

```python
records_n = len(train_split) + len(valid_split) + len(test_split)
print("Actual ratio is: {:2f}/{}/{}".format( len(train_split)/records_n, len(valid_split)/records_n, len(test_split)/records_n))
```

```python
brats_train = train_split['path'].to_list()
brats_valid = valid_split['path'].to_list()
brats_test = test_split['path'].to_list()

train_split.to_csv('../datasets/meta/brats2019v2_training.csv')
valid_split.to_csv('../datasets/meta/brats2019v2_validation.csv')
test_split.to_csv('../datasets/meta/brats2019v2_testing.csv')
```

```python
# Pipeline for preparing BRATS2019 for SegAN (Cropped)
center_crop = [180,180,128]
#brats_train, brats_valid = dh.split_train_validation(['../../datasets/BRATS2019/MICCAI_BraTS_2019_Data_Training/HGG/*/', '../../datasets/BRATS2019/MICCAI_BraTS_2019_Data_Training/LGG/*/'], ratio_train=0.9)

train_min, train_max = dh.find_extreme_values(file_paths=brats_train, pattern='/*.nii.gz', center_crop=center_crop)
valid_min, valid_max = dh.find_extreme_values(file_paths=brats_valid, pattern='/*.nii.gz', center_crop=center_crop)
test_min, test_max = dh.find_extreme_values(file_paths=brats_test, pattern='/*.nii.gz', center_crop=center_crop)
```

```python
dh.pack_dataset(brats_train, 'brats2019_v2', dataset_suffix='training', dataset_min=train_min, dataset_max=train_max, center_crop=center_crop, normalize_to='mri')
dh.pack_dataset(brats_valid, 'brats2019_v2', dataset_suffix='validation', dataset_min=valid_min, dataset_max=valid_max, center_crop=center_crop, normalize_to='mri')
dh.pack_dataset(brats_test, 'brats2019_v2', dataset_suffix='testing', dataset_min=test_min, dataset_max=test_max, center_crop=center_crop, normalize_to='mri')
```
